## **Course Objectives**

1. **Partial Differential Equations (PDEs)**
   - Gain an understanding of partial differential equations, including their formation and solution.

2. **Complex-Valued Functions**
   - Understand the concept of complex-valued functions and analytic functions.
   - Apply these concepts to solve real-life problems.

## **Course Outcomes**

Upon successful completion of this course, students will be able to:

1. **Apply PDEs to Real-World Problems**
   - Formulate and solve partial differential equations relevant to various scientific and engineering fields.

2. **Analyze Complex-Valued Functions**
   - Demonstrate an understanding of complex-valued functions and their applications.
   - Utilize analytic functions to address practical problems.

## **Module 1: Partial Differential Equations (PDEs)**

### **Learning Outcomes**
- Understand different types of PDEs and their classifications.
- Formulate PDEs from physical scenarios and solve them using appropriate methods.


### **Overview**
- **Introduction to PDEs**
  - Definition and examples of PDEs.
  - Difference between ordinary differential equations (ODEs) and PDEs.
  
- **Types of PDEs**
  - Classification: elliptic, parabolic, and hyperbolic PDEs.
  
- **Formation and Solution Methods**
  - How to formulate PDEs from real-world problems.
  - Common methods for solving PDEs: separation of variables, method of characteristics.

- **Examples and Applications**
  - Heat equation, wave equation, and Laplace’s equation.
  - Physical interpretations and boundary conditions.

## **Module 3: Complex Variable – Differentiation**

### **Overview**
- **Introduction to Complex Functions**
  - Definition of complex functions and their properties.
  
- **Analytic Functions**
  - Definition and conditions for a function to be analytic.
  - Cauchy-Riemann equations and their significance.

- **Differentiation in the Complex Plane**
  - Techniques for differentiating complex functions.
  - Examples of complex differentiable functions and their applications.

### **Learning Outcomes**
- Understand the concept of complex differentiation.
- Apply differentiation techniques to complex functions.

## **Module 5: Complex Variable – Residue Integration**

### **Overview**
- **Residue Theory**
  - Definition of residues and their computation.
  
- **Residue Theorem**
  - Statement and proof of the Residue Theorem.
  - Applications of residue theory to evaluate integrals.

- **Practical Examples**
  - Using residues to solve real integrals and analyze complex functions.

### **Learning Outcomes**
- Understand and calculate residues for complex functions.
- Apply residue theory to evaluate complex integrals.
